#### cleaning the data

```
## removing characters from the column names of the dataframe for easier access
cols = ['_'.join(re.sub('[()-_/:]', ' ', name).split()) for name in df.columns]
df.columns = cols

# also we don't need the first column, lets remove it
df.drop(['nnamed'], axis=1, inplace=True)

# not too many na's, so I will just remove them all from the data
df.dropna(inplace=True)
```

```
df.head()
```

```
# now separating the columns when they have dashes in the data 
# this involves getting the columns that contain this format and then splitting the columns
dashed_cols = [col for col in df.columns if isinstance(df.loc[0, col], str) and '-' in df.loc[0, col]]
```

```
def add_dashed_cols(col: str, df: pd.DataFrame):
    """
    Function to take a column from the df, split the data into two parts, and append it to the df. 
    :param col: column from the DataFrame
    :param df: DataFrame where the data is contained
    :returns: DataFrame with the new columns appended to the end and the column passed is dropped from the DataFrame
    """
    new_col_names = np.array([col+'first_number', col+'second_number'])

    def string_splitter(text: str, location: int) -> str:
        """
        Function to split the string in the column, and then acces the element with the location parameter. 
        If the split results in a single value of nan, the value 0 is returned meaning that the stat did not exist.
        :param text: string in the specific column and row
        :param location: either the 0 or 1 location in the split list
        :returns: the value in the split list at the specified location
        """
        if text == 'nan':
            return '0'
        return text.split('-')[location]

    temp_data = df[col].to_numpy(dtype=str)
    first_entry = np.vectorize(string_splitter)(temp_data, 0)
    second_entry = np.vectorize(string_splitter)(temp_data, 1)

    df[new_col_names[0]] = first_entry
    df[new_col_names[1]] = second_entry

    return df.drop([col], axis=1)

```

```
# reformatting the data to remove stats that use the form 4 of 5 in the form 4-5 in the column
# this means removing the dash and making an individual column for the first value and the second value
for col in dashed_cols:
    df = add_dashed_cols(col, df)
```

```
def total_time(possession_time: str) -> int:
    """
    Function to convert the time passed in from a string to the total seconds of the possession.
    :param possession_time: time as minutes:seconds as a string
    :returns: int format of the total time of possession
    """
    minutes, seconds = possession_time.split(':')
    return int(minutes) * 60 + int(seconds)
```

```
def apply_col(base_list, suffix, data, func) -> None:
    for team in base_list:
        col = team + suffix 
        data[col] = data[col].map(func)
```

```
# now we are going to change the format of the time of posession to total seconds since it is not a date time
home_away_list = ['home', 'away']
apply_col(home_away_list, '_possession', df, total_time)
```

```
# get rid of the scores and add in the win column (1 for a home win and 0 for an away win)
temp_series = df['home_points'].ge(df['away_points'])
df['home_win'] = temp_series.map(lambda x: 1 if x else 0)
df.drop(['home_points', 'away_points'], axis=1, inplace=True)
```
